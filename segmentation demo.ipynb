```python
from sem import SEM
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

opts = {'lambda': 10, 'alpha': 0.1, 'beta': 0.00001, 'eta': 0.01, 'tau': 0.1}

s = SEM(opts)

X = [[1, 1]] * 10 + [[-1, -1]] * 10

e, post = s.segment(X)
```

```python
K = len(post[-1])
p = np.zeros((len(post), K));
events = [];
for i in range(len(post)):
    p[i,0:len(post[i])] = post[i]
    events.append('Event ' + str(i))
print p
```

```python
plt.plot(p)
plt.legend(events)
plt.xlabel('t')
plt.ylabel('P(e|s_{1:t})')
plt.show()
```
